# QUESTION 1(a). Task: LSTM For Sequence Classification with Dropout

```python
# imports
import numpy
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Embedding
from keras.layers import Dropout
from keras.preprocessing import sequence
import matplotlib.pyplot as plt
```

```python
# fix random seed for reproducibility
numpy.random.seed(7)
```

```python
# load the dataset but only keep the top n words, zero the rest
top_words = 5000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=top_words)
```

```python
# truncate and pad input sequences
max_review_length = 500
X_train = sequence.pad_sequences(X_train, maxlen=max_review_length)
X_test = sequence.pad_sequences(X_test, maxlen=max_review_length)
```

```python
# fixed variable that does not change in question; declared as a global constant
embedding_vecor_length = 32
```

```python
# compile and train model
def train_model(model):
  history=model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3, batch_size=64)
  return history
```

```python
# Final evaluation of the model
def evaluate_model(model):
  scores = model.evaluate(X_test, y_test, verbose=0)
  print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
def plot_history(history):
  plt.figure(figsize=(4,4))
  plt.plot(history.history['loss'], label='train')
  plt.plot(history.history['val_loss'], label='test')
  plt.legend()
  plt.show()
```

## Test the LSTM with 2 kinds of dropout

LSTM is originally defined as:

model = Sequential()

model.add(Embedding(top_words, embedding_vecor_length, input_length=max_review_length))

model.add(LSTM(100))

model.add(Dense(1, activation= sigmoid ))

### We can do this easily by adding new Dropout layers between the Embedding and LSTM layers and the LSTM and Dense output layers.

```python
# create the model with new Dropout layers btwn Embedding & LSTM layers and LSTM & Dense output layers
model1 = Sequential()
model1.add(Embedding(top_words, embedding_vecor_length, input_length=max_review_length))
model1.add(Dropout(0.2))
model1.add(LSTM(100))
model1.add(Dropout(0.2))
model1.add(Dense(1, activation='sigmoid'))
```

```python
model1.build(input_shape=(None, max_review_length))
# compile model
model1.compile(loss= 'binary_crossentropy' , optimizer= 'adam' , metrics=[ 'accuracy' ]) #binary classification problem
print(model1.summary())
```

```python
# train model
history1=train_model(model=model1)
```

```python
# plot history
plot_history(history1)
```

```python
# evaluate model
evaluate_model(model1)
```

### We can also add dropout directly to the input on the Embedded layer by using the dropout parameter.



There is no dropout parameter in Embedding layer so Dropout was experimented by using LSTM's dropout parameter instead.

```python
# create the model
model2 = Sequential()
model2.add(Embedding(top_words, embedding_vecor_length, input_length=max_review_length))
model2.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model2.add(Dense(1, activation='sigmoid'))
```

```python
model2.build(input_shape=(None, max_review_length))
# compile model
model2.compile(loss= 'binary_crossentropy' , optimizer= 'adam' , metrics=[ 'accuracy' ]) #binary classification problem
print(model2.summary())
```

```python
# train model
history2=train_model(model=model2)
```

```python
# plot history
plot_history(history=history2)
```

```python
# evaluate model
evaluate_model(model=model2)
```

## COMMENTS

In the sequence classification task on the IMDB dataset, the original LSTM model without dropout achieved an accuracy of 86.79%. Applying dropout with separate layers between the embedding and LSTM layers, and between the LSTM and output layers, resulted in an accuracy of 85.82%. This approach effectively reduced overfitting and improved generalization. In contrast, integrating dropout directly within the LSTM layer yielded a lower accuracy of 82.31%.

While dropout generally aids in regularization, direct application within the LSTM may introduce noise and be less effective compared to using dropout layers strategically placed before and after the LSTM.

Separate dropout layers offer better regularization and performance compared to dropout within LSTM layers, but the original model without dropout still outperforms both, suggesting that dropout should be carefully tuned to balance regularization and model performance.

# QUESTION 1(b). Task: LSTM and CNN For Sequence Classification

The CNN-LSTM model was implemented to enhance sentiment classification of IMDB reviews by combining convolutional and recurrent layers. It starts with an embedding layer, followed by a 1D convolutional layer with 32 filters and a kernel size of 3 to extract local spatial features. A max pooling layer with a pool size of 2 then reduces the feature map size before passing the output to an LSTM layer with 100 units to capture sequential dependencies. The model concludes with a dense layer with a sigmoid activation function for binary classification.

## Create this CNN-LSTM architecture and test it on the IMDB data.

```python
# additional imports for 1D CNN & max pooling layer
from keras.layers import Conv1D
from keras.layers import MaxPooling1D
```

```python
# create model
model3 = Sequential()
model3.add(Embedding(top_words, embedding_vecor_length, input_length=max_review_length))
model3.add(Conv1D(filters=32, kernel_size=3, padding='same', activation='relu'))
model3.add(MaxPooling1D(pool_size=2))
model3.add(LSTM(100))
model3.add(Dense(1, activation='sigmoid'))
```

```python
model3.build(input_shape=(None, max_review_length))
# compile model
model3.compile(loss= 'binary_crossentropy' , optimizer= 'adam' , metrics=[ 'accuracy' ]) #binary classification problem
print(model3.summary())
```

```python
# train model
history3=train_model(model=model3)
```

```python
# plot history
plot_history(history=history3)
```

```python
# evaluate model
evaluate_model(model=model3)
```

## COMMENTS

The model achieved an impressive accuracy of 88.33%, demonstrating strong performance in classifying movie reviews into positive or negative sentiment.

 The CNN-LSTM architecture significantly outperforms the standalone LSTM model, highlighting the advantage of integrating convolutional layers for enhanced feature extraction followed by sequence modeling.

# QUESTION 2(a).

Imports

```python
import numpy as np
from keras.datasets import cifar10
from keras.layers import Input, Conv2D, BatchNormalization, ReLU, Add, AveragePooling2D, Flatten, Dense, Activation, GlobalAveragePooling2D
from keras.models import Model
from keras.utils import plot_model, to_categorical
from keras.optimizers import Adam
import matplotlib.pyplot as plt
from tensorflow.keras.applications import ResNet50
```

Load dataset

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

Residual Block

```python
def resnet_layer(inputs, num_filters=16, kernel_size=3, strides=1, activation=True, batch_normalization=True):
    x = Conv2D(num_filters, kernel_size=kernel_size, strides=strides, padding='same', use_bias=False)(inputs)
    if batch_normalization:
        x = BatchNormalization()(x)
    if activation:
        x = ReLU()(x)
    return x
```

Resnet Version 1 Model for Cifar10

```python
def resnet_v1(input_shape, depth, num_classes=10):
    """ResNet Version 1 Model builder [a]
      Stacks of 2 x (3 x 3) Conv2D-BN-ReLU
      Last ReLU is after the shortcut connection.
      At the beginning of each stage, the feature map size is halved
      (downsampled) by a convolutional layer with strides=2, while
      the number of filters is doubled. Within each stage,
      the layers have the same number filters and the
      same number of filters.
      Features maps sizes:
      stage 0: 32x32, 16
      stage 1: 16x16, 32
      stage 2: 8x8, 64
      The Number of parameters is approx the same as Table 6 of [a]:
      ResNet20 0.27M
      ResNet32 0.46M
      ResNet44 0.66M
      ResNet56 0.85M
      ResNet110 1.7M
      Arguments:
      input_shape (tensor): shape of input image tensor
      depth (int): number of core convolutional layers
      num_classes (int): number of classes (CIFAR10 has 10)
      Returns:
      model (Model): Keras model instance
      """

    if (depth - 2) % 6 != 0:
      raise ValueError('depth should be 6n+2 (eg 20, 32, in [a])')
    # Start model definition.
    num_filters = 16
    num_res_blocks = int((depth - 2) / 6)
    inputs = Input(shape=input_shape)
    x = resnet_layer(inputs=inputs)
    # instantiate the stack of residual units
    for stack in range(3):
      for res_block in range(num_res_blocks):
        strides = 1
        # first layer but not first stack
        if stack > 0 and res_block == 0:
          strides = 2 # downsample
        y = resnet_layer(inputs=x,
        num_filters=num_filters,
        strides=strides)
        y = resnet_layer(inputs=y,
        num_filters=num_filters,
        activation=None)
        # first layer but not first stack
        if stack > 0 and res_block == 0:
    # linear projection residual shortcut
    # connection to match changed dims
          x = resnet_layer(inputs=x,
          num_filters=num_filters,
          kernel_size=1,
          strides=strides,
          activation=None,
          batch_normalization=False)
          x = Add()([x, y])
          x = Activation('relu')(x)
        num_filters *= 2
    # add classifier on top.
    # v1 does not use BN after last shortcut connection-ReLU
    x = AveragePooling2D(pool_size=8)(x)
    y = Flatten()(x)
    outputs = Dense(num_classes,
              activation='softmax',
              kernel_initializer='he_normal')(y)
    # instantiate model.
    model = Model(inputs=inputs, outputs=outputs)
    return model
```

ResNet50 Model

```python
resnet50_model = ResNet50(weights='imagenet',include_top=False, input_shape=(32, 32, 3))
```

ResNet20 Model

```python
resnet20_model = resnet_v1(input_shape=(32, 32, 3), depth=20)
```

Compiling the ResNet Models

```python
# Compile ResNet-20 model
resnet20_model.compile(optimizer=Adam(learning_rate=0.001),
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])
print(resnet20_model.summary())
```

```python
# Plot ResNet-20
plot_model(resnet20_model, to_file='resnet20_model.png', show_shapes=True, show_layer_names=True)
```

```python
# compile ResNet-50 model
x = resnet50_model.output
x = GlobalAveragePooling2D()(x)
outputs = Dense(10, activation='softmax')(x)

resnet50 = Model(inputs=resnet50_model.input, outputs=outputs)

# Compile the model
resnet50.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
print(resnet50.summary())
```

Training the ResNet Models

```python
from keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# Train ResNet-20
history_resnet20 = resnet20_model.fit(x_train, y_train,
                                epochs=50,
                                batch_size=64,
                                validation_split=0.1,
                                callbacks=[early_stopping])
```

```python
# Train ResNet-50
history_resnet50 = resnet50.fit(x_train, y_train,
                                epochs=50,
                                batch_size=64,
                                validation_split=0.1,
                                callbacks=[early_stopping])
```

Evaluating the ResNet Models

```python
# Evaluate ResNet-20
loss_resnet20, accuracy_resnet20 = resnet20_model.evaluate(x_test, y_test, verbose=0)
print(f"ResNet-20 Test Loss: {loss_resnet20:.4f}")
print(f"ResNet-20 Test Accuracy: {accuracy_resnet20:.4f}")
```

```python
# Evaluate ResNet-50
loss_resnet50, accuracy_resnet50 = resnet50.evaluate(x_test, y_test, verbose=0)
print(f"ResNet-50 Test Loss: {loss_resnet50:.4f}")
print(f"ResNet-50 Test Accuracy: {accuracy_resnet50:.4f}")
```

Plot training & validation accuracy values

```python
# ResNet-20
plt.figure(figsize=(12, 5))
plt.plot(history_resnet20.history['accuracy'])
plt.plot(history_resnet20.history['val_accuracy'])
plt.title('ResNet-20 model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'])
plt.show()
```

```python
# ResNet-50
plt.figure(figsize=(12, 5))
plt.plot(history_resnet50.history['accuracy'])
plt.plot(history_resnet50.history['val_accuracy'])
plt.title('ResNet-50 model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'])
plt.show()
```

## COMMENTS

The CIFAR-10 dataset, known for its smaller image size and complexity, is prone to overfitting, especially with deeper networks like ResNet-50. ResNet-20, with its simpler architecture, demonstrated better generalization, achieving a test accuracy of 83.83% and a low test loss of 0.7780. In contrast, ResNet-50 performed well initially but struggled with high test loss (1.2035) and lower accuracy (61.87%) due to overfitting. ResNet-50's training curves indicated high accuracy and low loss initially but showed declining validation performance over time. This suggests that while ResNet-50 is more powerful, it requires more sophisticated regularization to avoid overfitting compared to ResNet-20.
